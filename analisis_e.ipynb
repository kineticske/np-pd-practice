Generar un DataFrame con los datos de los cuatro ficheros.

```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv(r'emisiones-2016.csv', sep=';')
data2=pd.read_csv(r'emisiones-2017.csv', sep=';')
data3=pd.read_csv(r'emisiones-2018.csv', sep=';')
data4=pd.read_csv(r'emisiones-2019.csv', sep=';')
dataT=pd.concat([data1, data2, data3, data4])
dataT
```

Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÃ‘O, MES y las correspondientes a los dÃ­as D01, D02, etc.

```python
columnas=['ESTACION', 'MAGNITUD', 'ANO', 'MES']
columnas.extend([col for col in dataT if col.startswith('D')])

dataT=dataT[columnas]
dataT
```

Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los dÃ­as aparezcan en una Ãºnica columna.

```python
dataT=dataT.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='VALOR') #columnas en un valor y los valores
dataT
```

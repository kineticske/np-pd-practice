Generar un DataFrame con los datos de los cuatro ficheros.

```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
data1=pd.read_csv(r'emisiones-2016.csv', sep=';')
data2=pd.read_csv(r'emisiones-2017.csv', sep=';')
data3=pd.read_csv(r'emisiones-2018.csv', sep=';')
data4=pd.read_csv(r'emisiones-2019.csv', sep=';')
dataT=pd.concat([data1, data2, data3, data4])
dataT
```

Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÃ‘O, MES y las correspondientes a los dÃ­as D01, D02, etc.

```python
columnas=['ESTACION', 'MAGNITUD', 'ANO', 'MES']
columnas.extend([col for col in dataT if col.startswith('D')])

dataT=dataT[columnas]
dataT
```

Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los dÃ­as aparezcan en una Ãºnica columna.

```python
dataT=dataT.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='VALOR') #columnas en un valor y los valores
dataT
```

Crear una nueva columna con las fechas a partir del aÃ±o, mes y dÃ­a

```python
dataT['DIA']=dataT.DIA.str.strip('D') #The strip() method removes any leading (spaces at the beginning) and trailing (spaces at the end) characters (space is the default leading character to remove)
dataT['FECHA']=dataT['DIA'].apply(str) + '/' + dataT['MES'].apply(str) + '/' + dataT['DIA'].apply(str)
dataT['FECHA']=pd.to_datetime(dataT['FECHA'], format='%Y/%m/%d', infer_datetime_format=True, errors='coerce')

dataT
```

```python
dataT[(dataT.ESTACION==4)&(dataT.MAGNITUD==1)].groupby('ANO').mean()
```

Eliminar las filas con fechas no vÃ¡lidas (utilizar la funciÃ³n isnat del mÃ³dulo numpy) y ordenar el DataFrame por estaciones, contaminantes y fecha.

```python
dataT=dataT.drop(dataT[np.isnat(dataT.FECHA)].index)
dataT
```

```python
#ordenando valores
dataT.sort_values(['ESTACION', 'MAGNITUD', 'FECHA'])
```

Crear una funciÃ³n que reciba una estaciÃ³n, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estaciÃ³n y rango de fechas dado.

```python
def meteo(estacion, contaminante, desde, hasta):
    return dataT[(dataT.ESTACION==estacion) & (dataT.MAGNITUD==contaminante) & (dataT.FECHA>=dt.datetime.strptime(desde, '%Y/%m/%d'))& (dataT.FECHA<=dt.datetime.strptime(hasta, '%Y/%m/%d'))]

meteo(4,1, '2001/01/01','2001/01/02')
```

Crear una funciÃ³n que reciba una estaciÃ³n y un contaminante y devuelva un resumen descriptivo de las emisiones del contaminante indicado en la estadiÃ³n indicada.

```python
def describeme(estacion, contaminante):
    return dataT[(dataT.ESTACION==estacion) & (dataT.MAGNITUD==contaminante)].VALOR.describe()
#dataT[(dataT.ESTACION==estacion) & (dataT.MAGNITUD==contaminante)]
describeme(4,1)
```

Crear una funciÃ³n que devuelva las emisiones medias mensuales de un contaminante y un aÃ±o dados para todos las estaciones.

```python
def emisiones_medias(contaminante, anio):
    return dataT[(dataT.MAGNITUD==contaminante) & (dataT.ANO==anio)].groupby(['ESTACION', 'MES']).VALOR.mean().unstack()

emisiones_medias(1,2016)
```

Crear un funciÃ³n que reciba una estaciÃ³n de mediciÃ³n y devuelva un DataFrame con las medias mensuales de los distintos tipos de contaminantes.

```python
def mediasContaminantes(estacion):
    return dataT[dataT.ESTACION==estacion].groupby(['MAGNITUD','MES']).VALOR.mean().unstack()

mediasContaminantes(4)
```
